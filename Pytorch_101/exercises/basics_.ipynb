### Create a straight line dataset using the linear regression formula (weight * X + bias).
* Set weight=0.3 and bias=0.9 there should be at least 100 datapoints total.
* Split the data into 80% training, 20% testing.
*Plot the training and testing data so it becomes visual.

```python
import torch as th
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = th.linspace(0,10,100).reshape(100,)
weight = 0.3
bias = 0.9
Y = weight * X + bias
dataset = th.stack([X,Y],1)

# Splitting The dataset
train_set, test_set = dataset[:int(len(dataset)*0.8),:], dataset[int(len(dataset)*0.8):,:]


# Plotting the Dataset
fig,ax = plt.subplots()
ax.scatter(train_set[:,0],train_set[:,1],c='red')
ax.scatter(test_set[:,0],test_set[:,1],c='blue')
```
